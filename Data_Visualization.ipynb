```python
numerical_summary = df[['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']].agg(['mean', 'median', 'count'])
print(numerical_summary)
```

```python
product_summary = df.groupby('Product').agg(
    Total_Revenue=('TotalPrice', 'sum'),
    Average_Spend=('TotalPrice', 'mean'),
    Total_Quantity=('Quantity', 'sum'),
    Order_Count=('OrderID', 'count')
).sort_values(by='Total_Revenue', ascending=False)
print(product_summary)
```

```python
referral_summary = df.groupby('ReferralSource')['TotalPrice'].sum().sort_values(ascending=False)
print(referral_summary)
```

```python
monthly_trends = df.groupby('YearMonth').agg(
    Monthly_Revenue=('TotalPrice', 'sum'),
    Order_Volume=('OrderID', 'count')
).sort_index()
```

```python
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['TotalPrice'] < lower_bound) | (df['TotalPrice'] > upper_bound)]
```

```python
print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
print(f"Total number of outlier transactions found: {len(outliers)}")
print("\nTop 5 Highest Outlier Transactions:")
print(outliers[['OrderID', 'Product', 'Quantity', 'TotalPrice']].sort_values(by='TotalPrice', ascending=False).head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Order_Count'] = df['Product'].map(df['Product'].value_counts())

fig, ax1 = plt.subplots(figsize=(11, 6))

sns.barplot(x='Product', y='TotalPrice', data=df, ax=ax1, hue='Product')
ax1.set_title('Product Analysis: Total Revenue vs. Order Volume', fontweight='bold', pad=20)
ax1.set_xlabel('Product Category', fontweight='bold')
ax1.set_ylabel('Total Revenue ($)', fontweight='bold')

ax2 = ax1.twinx()
sns.lineplot(x='Product', y='Order_Count', data=df, ax=ax2, marker='o', linewidth=3, markersize=8)
ax2.set_ylabel('Total Orders Placed')
ax2.grid(False)

plt.tight_layout()
plt.savefig('product_analysis.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(9, 5))

sns.barplot(x='TotalPrice', y='ReferralSource', data=df, palette='rocket', hue='ReferralSource')
plt.title('Acquisition Efficiency: Revenue Generated by Referral Source')
plt.xlabel('Total Revenue Generated ($)')
plt.ylabel('Referral Marketing Channel')

plt.tight_layout()
plt.savefig('referral_efficiency.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Date'] = pd.to_datetime(df['Date'])
df['YearMonth'] = df['Date'].dt.to_period('M').astype(str)

df_monthly = df.groupby('YearMonth').agg(
    TotalPrice=('TotalPrice', 'sum'),
    Order_Count=('OrderID', 'count')
).reset_index().sort_values('YearMonth')

fig, ax1 = plt.subplots(figsize=(13, 6))

sns.lineplot(x='YearMonth', y='TotalPrice', data=df_monthly, ax=ax1, marker='s', linewidth=2.5, markersize=7, color='tab:blue')
ax1.set_title('Macro Trends: Core Financial Performance (Jan - Oct 2023)', fontweight='bold', pad=20, fontsize=14)
ax1.set_xlabel('Timeline (Year-Month)', fontweight='bold')
ax1.set_ylabel('Monthly Performance Revenue ($)', fontweight='bold', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax1.set_xticks(range(len(df_monthly)))
ax1.set_xticklabels(df_monthly['YearMonth'], rotation=30)

ax2 = ax1.twinx()
sns.lineplot(x='YearMonth', y='Order_Count', data=df_monthly, ax=ax2, marker='^', linewidth=2.5, markersize=7, color='tab:orange')
ax2.set_ylabel('Monthly Transaction Volume', fontweight='bold', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')
ax2.grid(False)

ax1.annotate('Peak Season ($63.8K)', xy=(4, 63836.84), xytext=(5, 62000),
             arrowprops=dict(facecolor='darkgreen', shrink=0.08, width=1, headwidth=6),
             fontweight='bold', color='darkgreen')

ax1.annotate('Q2 Revenue Slump', xy=(3, 27751.71), xytext=(1.2, 26000),
             arrowprops=dict(facecolor='crimson', shrink=0.08, width=1, headwidth=6),
             fontweight='bold', color='crimson')

plt.tight_layout()
plt.savefig('macro_trends_seasonality.png', dpi=300)
plt.show()
```
